```python
class Atm:
    # constructor is function inside the class. it is a special function
    def __init__(self):
        self.pin = ''
        self.balance = 0
        self.menu()

    def menu(self):
        user_input = input("""
Hi how can I help you?
1. Press 1 to create pin
2. Press 2 to change pin
3. Press 3 to check balance
4. Press 4 to withdraw
5. Anything else to exit
""")

        if user_input == "1":
            self.create_pin()

        elif user_input == "2":
            self.change_pin()

        elif user_input == "3":
            self.check_balance()

        elif user_input == "4":
            self.withdraw()

        else:
            exit()

    def create_pin(self):
        user_pin = input("Enter your pin: ")
        self.pin = user_pin

        user_balance = input("Enter your balance: ")
        self.balance = int(user_balance)
        print("Pin created successfully")
        self.menu()
        
    def change_pin(self):
        old_pin=input('enter old pin')
        if old_pin==self.pin:
            new_pin=input('enter new pin')
            self.pin=new_pin
            print('pin changed sucessfully')
            self.menu()
        else:
            print("not vaild ")
        self.menu()
    def check_balance(self):
        input('enter your pin')
        if user_pin == self.pin:
            print('your balance is',self.balance)
        else:
            print("invalid or wrong pin")
    def withdraw(self):
        user_pin=input("enter the pin")
        if user_pin == self.pin:
            amount=int(input('enter your amount'))
            if amount<=self.balance:
                self.balance=self.balance-amount
                print('withdrawl successful balance is',self.balance)
            else:
                print("incorrect amount please enter less amount")
        else:
            print("error ")
            self.menu()
        
```

```python
#parametrized constructor
class Fraction:
    def __init__(self,x,y):
        self.num=x
        self.den=y
    def __str__(self):
        return'{}/{}'.format(self.num,self.den)
    def __add__(self,other):
        new_num=self.num*other.den + other.num*self.den
        new_dem=self.den*other.den
        return'{}/{}'.format(new_num,new_dem)
    def __sub__(self,other):
        new_num=self.num*other.den - other.num*self.den
        new_dem=self.den*other.den
        return'{}/{}'.format(new_num,new_dem)     
    def __mul__(self,other):
        new_num=self.num*other.num
        new_dem=self.den*other.den
        return'{}/{}'.format(new_num,new_dem)
    def __truediv__(self,other):
        new_num=self.num*other.den 
        new_dem=self.den*other.num
        return'{}/{}'.format(new_num,new_dem)
```

```python
fr1=Fraction(3,4)
fr2=Fraction(7,14)
```

```python
print(fr1)
print(fr2)
```
